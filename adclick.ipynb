```python
import pandas as pd
ad_clicks = pd.read_csv('ad_clicks.csv')
```

```python
ad_clicks.head()
```

```python
ad_clicks.groupby('utm_source').user_id.count().reset_index()
```

```python
ad_clicks['is_click'] = ad_clicks.ad_click_timestamp.isnull()
ad_clicks.is_click
```

```python
clicks_by_source = ad_clicks.groupby(['utm_source','is_click']).user_id.count().reset_index()
clicks_by_source
```

```python
clicks_pivot = clicks_by_source.pivot(columns='is_click',index='utm_source',values='user_id')
clicks_pivot
```

```python
clicks_pivot['percent_clicked'] = clicks_pivot[True]/(clicks_pivot[True]+clicks_pivot[False])
clicks_pivot.percent_clicked
```

```python
ad_clicks.groupby('experimental_group').user_id.count().reset_index()
```

```python
ad_clicks.groupby(['experimental_group','is_click']).user_id.count().reset_index()
```

```python
a_clicks = ad_clicks[ad_clicks.experimental_group == 'A']
b_clicks = ad_clicks[ad_clicks.experimental_group == 'B']
b_clicks
```

```python
x = a_clicks.groupby(['is_click','day']).user_id.count().reset_index()
x
```

```python
y = x.pivot(columns='is_click',index='day',values='user_id').reset_index()
y
```

```python
y['percent_clicked'] = y[True]/y[True]+y[False]
y.percent_clicked
```

```python
c = b_clicks.groupby(['is_click','day']).user_id.count().reset_index()
c
```

```python
d = c.pivot(columns='is_click',index='day',values='user_id').reset_index()
d
```

```python
d['percent_clicked'] = d[True]/d[True]+d[False]
d.percent_clicked
```
